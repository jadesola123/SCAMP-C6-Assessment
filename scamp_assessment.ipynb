```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
url='https://raw.githubusercontent.com/she-code-africa/SCAMP-Cohort6-Technical-Assessment/master/intermediate/fake_job_postings.csv'
jobs = pd.read_csv(url)

jobs.dropna(subset = ["industry"], inplace=True)
print(jobs.shape)
```

```python
## to create a new row for counting the fraudulence in each industry

fraud_jobs= jobs.groupby('industry').agg({'fraudulent':sum})

print(fraud_jobs.tail(50))
```

The industry with the highest fraud is Oil and Energy

```python
fraud_jobs.plot(legend=True)
plt.xticks(rotation=90)
plt.show()
```
